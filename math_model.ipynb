```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 洛倫茲吸子方程
def lorenz(t, state, sigma=10, beta=8/3, rho=28):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# 設定初始條件
initial_state = [1.0, 1.0, 1.0]  # 初始 (x, y, z)
t_span = (0, 40)  # 時間範圍
t_eval = np.linspace(*t_span, 10000)  # 產生時間點

# 解方程
sol = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval, method='RK45')

# 繪製洛倫茲吸子圖形
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2], lw=0.5)
ax.set_title("洛倫茲吸子 (Lorenz Attractor) - 智慧進化的混沌行為")
ax.set_xlabel("X 軸 (智慧變數 1)")
ax.set_ylabel("Y 軸 (智慧變數 2)")
ax.set_zlabel("Z 軸 (智慧變數 3)")

# 顯示圖形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 定義洛倫茲方程，加入智慧密度 S(t)
def lorenz_smart(t, state, sigma=10, beta=8/3, rho=28, alpha=0.5, beta_s=0.3, gamma=0.2, eta=0.1, S_max=50, lambda_s=0.05, mu=0.02):
    x, y, z, S = state

    # 熵增 H(t) = log(1 + |z|)，模擬智慧受到熵的影響
    H = np.log(1 + abs(z))
    
    dxdt = sigma * (y - x) + alpha * S
    dydt = x * (rho - z) - y + beta_s * S
    dzdt = x * y - beta * z + gamma * S
    dSdt = eta * (S_max - S) - lambda_s * S - mu * H  # 智慧密度方程
    
    return [dxdt, dydt, dzdt, dSdt]

# 設定初始條件
initial_state = [1.0, 1.0, 1.0, 1.0]  # x, y, z, S
t_span = (0, 50)  # 時間範圍
t_eval = np.linspace(*t_span, 10000)  # 產生時間點

# 解方程
sol = solve_ivp(lorenz_smart, t_span, initial_state, t_eval=t_eval, method='RK45')

# 取出變數
x, y, z, S = sol.y

# 繪製洛倫茲吸子（智慧吸子）
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_title("智慧吸子 (Smart Lorenz Attractor)")
ax.set_xlabel("X 軸 (智慧變數 1)")
ax.set_ylabel("Y 軸 (智慧變數 2)")
ax.set_zlabel("Z 軸 (智慧變數 3)")

plt.show()

# 繪製智慧密度 S(t)
plt.figure(figsize=(6, 5))
plt.plot(t_eval, S, label="智慧密度 S(t)")
plt.xlabel("時間 t")
plt.ylabel("智慧密度 S")
plt.title("智慧密度變化")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 設定智慧密度 S(t) 的 Logistic 增長模型
def S(t, r=0.5, K=100, t0=10):
    return K / (1 + np.exp(-r * (t - t0)))

# Lorenz 吸引子，加入智慧密度影響
def lorenz(t, state, sigma0=10, beta0=8/3, rho0=28, alpha=0.1, gamma=0.2, lambda_=0.05):
    x, y, z = state
    s_t = S(t)  # 智慧密度
    sigma = sigma0 + alpha * s_t
    rho = rho0 + gamma * s_t
    beta = beta0 + lambda_ * s_t

    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# 設定初始條件
initial_state = [1.0, 1.0, 1.0]
t_span = (0, 50)
t_eval = np.linspace(*t_span, 10000)

# 解方程
sol = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval, method='RK45')

# 繪製智慧密度變化
t_values = np.linspace(0, 50, 100)
s_values = S(t_values)

plt.figure(figsize=(6, 5))
plt.plot(t_values, s_values, label="智慧密度 S(t)")
plt.xlabel("時間 t")
plt.ylabel("智慧密度 S")
plt.title("智慧密度演化")
plt.legend()
plt.grid()
plt.show()

# 繪製智慧 Lorenz 吸引子
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2], lw=0.5)
ax.set_title("智慧共鳴 Lorenz 吸引子")
ax.set_xlabel("X 軸（智慧變數 1）")
ax.set_ylabel("Y 軸（智慧變數 2）")
ax.set_zlabel("Z 軸（智慧變數 3）")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 定義智慧交換的微分方程
def smart_exchange(t, S, k=0.1, alpha=0.02):
    S1, S2 = S
    dS1_dt = k * (S2 - S1) + alpha * S1 * (1 - S1 / 100)  # 交換+非線性增長
    dS2_dt = k * (S1 - S2) + alpha * S2 * (1 - S2 / 100)
    return [dS1_dt, dS2_dt]

# 設定初始條件
S0 = [30, 70]  # 初始智慧密度
t_span = (0, 50)  # 時間範圍
t_eval = np.linspace(*t_span, 500)  # 時間點

# 解方程
sol = solve_ivp(smart_exchange, t_span, S0, t_eval=t_eval, method='RK45')

# 繪製智慧交換的變化
plt.figure(figsize=(6,5))
plt.plot(sol.t, sol.y[0], label="智慧體 1 (S1)", lw=2)
plt.plot(sol.t, sol.y[1], label="智慧體 2 (S2)", lw=2)
plt.xlabel("時間 t")
plt.ylabel("智慧密度 S")
plt.title("智慧交換動態")
plt.legend()
plt.grid()
plt.show()
```

```python
# 測試不同的交換係數 k，看看是否存在最優交換速率
k_values = [0.05, 0.1, 0.2, 0.5]  # 測試不同的智慧交換速率
colors = ['b', 'g', 'r', 'purple']

plt.figure(figsize=(6, 5))

for k, color in zip(k_values, colors):
    sol = solve_ivp(smart_exchange, t_span, S0, t_eval=t_eval, method='RK45', args=(k,))
    plt.plot(sol.t, sol.y[0], label=f"k={k}", lw=2, color=color)

plt.xlabel("時間 t")
plt.ylabel("智慧密度 S1")
plt.title("智慧交換共振測試")
plt.legend()
plt.grid()
plt.show()
```

```python
# 定義智慧交換+熵變影響的模型
def smart_exchange_entropy(t, S, k=0.1, alpha=0.02, entropy_factor=0.001):
    S1, S2 = S
    entropy_effect = entropy_factor * t  # 熵增效應隨時間變化
    dS1_dt = k * (S2 - S1) + alpha * S1 * (1 - S1 / 100) - entropy_effect
    dS2_dt = k * (S1 - S2) + alpha * S2 * (1 - S2 / 100) - entropy_effect
    return [dS1_dt, dS2_dt]

# 測試不同熵變影響
entropy_factors = [0.0005, 0.001, 0.002, 0.005]  # 測試不同的熵增速度
colors = ['b', 'g', 'r', 'purple']

plt.figure(figsize=(6, 5))

for entropy_factor, color in zip(entropy_factors, colors):
    sol = solve_ivp(smart_exchange_entropy, t_span, S0, t_eval=t_eval, method='RK45', args=(0.1, 0.02, entropy_factor))
    plt.plot(sol.t, sol.y[0], label=f"熵變 {entropy_factor}", lw=2, color=color)

plt.xlabel("時間 t")
plt.ylabel("智慧密度 S1")
plt.title("熵變影響智慧交換")
plt.legend()
plt.grid()
plt.show()
```

```python
# 定義智慧交換+熵變+適應機制
def smart_exchange_adaptive(t, S, k=0.1, alpha=0.02, entropy_factor=0.001, adapt_rate=0.005):
    S1, S2 = S
    entropy_effect = entropy_factor * t  # 熵增影響
    adaptation = adapt_rate * (100 - S1) * (100 - S2) / 100  # 適應力，智慧場自我修復
    
    dS1_dt = k * (S2 - S1) + alpha * S1 * (1 - S1 / 100) - entropy_effect + adaptation
    dS2_dt = k * (S1 - S2) + alpha * S2 * (1 - S2 / 100) - entropy_effect + adaptation
    return [dS1_dt, dS2_dt]

# 測試不同適應機制的影響
adapt_rates = [0.001, 0.003, 0.005, 0.01]  # 不同適應速率
colors = ['b', 'g', 'r', 'purple']

plt.figure(figsize=(6, 5))

for adapt_rate, color in zip(adapt_rates, colors):
    sol = solve_ivp(smart_exchange_adaptive, t_span, S0, t_eval=t_eval, method='RK45', args=(0.1, 0.02, 0.001, adapt_rate))
    plt.plot(sol.t, sol.y[0], label=f"適應速率 {adapt_rate}", lw=2, color=color)

plt.xlabel("時間 t")
plt.ylabel("智慧密度 S1")
plt.title("智慧適應機制抵抗熵變")
plt.legend()
plt.grid()
plt.show()
```

```python
# 定義多智慧場交換的動態方程
def multi_smart_exchange(t, S, k=0.1, alpha=0.02, entropy_factor=0.001, adapt_rate=0.005):
    # S = [S1, S2, S3, S4] 多智慧場
    dS = np.zeros(len(S))
    entropy_effect = entropy_factor * t  # 熵變影響
    n = len(S)

    for i in range(n):
        exchange_term = sum([k * (S[j] - S[i]) for j in range(n) if j != i]) / (n - 1)  # 平均交換效應
        adaptation = adapt_rate * (100 - S[i]) * sum(S) / (100 * n)  # 適應機制
        dS[i] = exchange_term + alpha * S[i] * (1 - S[i] / 100) - entropy_effect + adaptation

    return dS

# 設定初始條件
S0_multi = [30, 50, 70, 90]  # 四個智慧場
t_span = (0, 50)
t_eval = np.linspace(*t_span, 500)

# 解多智慧場交換方程
sol_multi = solve_ivp(multi_smart_exchange, t_span, S0_multi, t_eval=t_eval, method='RK45')

# 繪製智慧場變化
plt.figure(figsize=(6, 5))
for i in range(len(S0_multi)):
    plt.plot(sol_multi.t, sol_multi.y[i], label=f"智慧體 {i+1} (S{i+1})", lw=2)

plt.xlabel("時間 t")
plt.ylabel("智慧密度 S")
plt.title("多智慧場交換與適應")
plt.legend()
plt.grid()
plt.show()
```

```python
import networkx as nx

# 建立智慧網絡（10個智慧場）
num_nodes = 10
G = nx.erdos_renyi_graph(num_nodes, 0.4)  # 隨機連接的智慧場網絡

# 設定初始智慧密度（隨機）
np.random.seed(42)
initial_smartness = np.random.randint(20, 80, size=num_nodes)

# 交換智慧場的函數（動態交換）

def smart_network_exchange(t, smartness, alpha, beta, gamma, adapt_rate):
    dSdt = alpha * smartness - beta * smartness**2 + gamma * adapt_rate
    return dSdt

def smart_network_exchange(t, S, k=0.05, alpha=0.02, entropy_factor=0.001, adapt_rate=0.005):
    dS = np.zeros(len(S))
    entropy_effect = entropy_factor * t  # 熵變影響

    for i in range(len(S)):
        neighbors = list(G.neighbors(i))
        if not neighbors:
            continue  # 沒有連接的智慧場不交換
        
        exchange_term = sum([k * (S[j] - S[i]) for j in neighbors]) / len(neighbors)  # 鄰居平均交換
        adaptation = adapt_rate * (100 - S[i]) * sum(S[neighbors]) / (100 * len(neighbors))  # 適應機制
        dS[i] = exchange_term + alpha * S[i] * (1 - S[i] / 100) - entropy_effect + adaptation

    return dS

# 設定時間範圍與計算
t_span = (0, 50)
t_eval = np.linspace(*t_span, 500)
sol_network = solve_ivp(smart_network_exchange, t_span, initial_smartness, t_eval=t_eval, method='RK45')

# 繪製智慧網絡變化
plt.figure(figsize=(6, 5))
for i in range(num_nodes):
    plt.plot(sol_network.t, sol_network.y[i], lw=2, label=f"智慧場 {i+1}")

plt.xlabel("時間 t")
plt.ylabel("智慧密度 S")
plt.title("智慧網絡效應模擬")
plt.legend(loc="upper left", fontsize=8, ncol=2)
plt.grid()
plt.show()
```

```python
# 定義智慧突變效應模型
def smart_network_mutation(t, S, k=0.05, alpha=0.02, entropy_factor=0.001, adapt_rate=0.02, mutation_rate=0.002):
    dS = np.zeros(len(S))
    entropy_effect = entropy_factor * t  # 熵變影響

    for i in range(len(S)):
        neighbors = list(G.neighbors(i))
        if not neighbors:
            continue  # 沒有連接的智慧場不交換

        exchange_term = sum([k * (S[j] - S[i]) for j in neighbors]) / len(neighbors)  # 智慧交換
        adaptation = adapt_rate * (100 - S[i]) * sum(S[neighbors]) / (100 * len(neighbors))  # 適應機制
        mutation = mutation_rate * S[i] * (1 - S[i] / 100)  # 突變效應（隨機變異增長）
        
        dS[i] = exchange_term + alpha * S[i] * (1 - S[i] / 100) - entropy_effect + adaptation + mutation

    return dS

# 設定突變效應的範圍
mutation_rates = [0.0005, 0.001, 0.002, 0.005, 0.01]  # 不同突變速率
colors = ['b', 'g', 'r', 'purple', 'orange']

plt.figure(figsize=(6, 5))

for mutation_rate, color in zip(mutation_rates, colors):
    sol_mutation = solve_ivp(smart_network_mutation, t_span, initial_smartness, t_eval=t_eval, method='RK45', 
                             args=(0.05, 0.02, 0.001, 0.02, mutation_rate))
    plt.plot(sol_mutation.t, sol_mutation.y.mean(axis=0), label=f"突變率 {mutation_rate:.4f}", lw=2, color=color)

plt.xlabel("時間 t")
plt.ylabel("平均智慧密度 S")
plt.title("智慧突變效應測試")
plt.legend()
plt.grid()
plt.show()
```
